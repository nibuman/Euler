If they are then placed in numerical order, with any repeats removed, we get the following sequence of 15 distinct terms:

4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 256, 625, 1024, 3125

How many distinct terms are in the sequence generated by $a^b$ for $2 ≤ a ≤ 100$ and $2 ≤ b ≤ 100$?


If $a^b = c$, and $d$ is a proper divisor of $b$, then:

\begin{equation}
  a^b = (a^d)^{\frac{b}{d}}
\end{equation}

e.g. for $3^8$ $d = 2$ or $4$. For $d = 2$:

\begin{equation*}
3^8 = 3\times 3\times 3\times 3\times 3\times 3\times 3\times 3 = 3^2\times 3^2\times 3^2\times 3^2 = 9\times 9\times 9\times 9 = 9^4
\end{equation*}
For $d = 4$ :
\begin{equation*}
3^8 = 3\times 3\times 3\times 3\times 3\times 3\times 3\times 3 = 3^4\times 3^4 = 81\times 81 = 81^2
\end{equation*}
